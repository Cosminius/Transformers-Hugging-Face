```python
from transformers import pipeline
from transformers.utils import logging

#seteaza nivelul de loguri doar la erori
logging.set_verbosity_error()
```

```python
#Sentiment analysis
classifier = pipeline("sentiment-analysis")

response = classifier("I feel happy and full of energy")
response2 = classifier(["I feel sad", "I feel happy"])
print(response)
print(response2)
```

```python
#Zero-shoot clasification

zsc_classifier = pipeline("zero-shot-classification")
zsc_response = zsc_classifier("I like to read books about Game of Thrones", candidate_labels=["fantasy", "war", "politics"])

print(zsc_response)
```

```python
#Test generation

tx_model = pipeline("text-generation")

tx_response = tx_model("This books is about chivalery and", max_length = 15 , num_return_sequences=2)

print(tx_response)
```

```python
#Use any model from hugging face

tx_model = pipeline("text-generation", model="Qwen/Qwen3-0.6B")

tx_response = tx_model("This is my name and is not enough")
print(tx_response)
```
